## 3.2 处理数据缺失

分别使用下列四种策略对缺失值进行处理:

1. 将缺失部分剔除
2. 用最高频率值来填补缺失值
3. 通过属性的相关关系来填补缺失值
4. 通过数据对象之间的相似性来填补缺失值

### - 缺失数据处理 records-for-2011.csv 

属性缺失值统计

```python
tmp_data = data[0]
print(tmp_data.isnull().sum(axis=0))
```

#### Area Id 属性缺失处理

缺失的原因可能是人为失误


- 将缺失部分剔除

```python
attri = "Area Id"
d = tmp_data.dropna(subset=[attri])
d
```

可视化对比

```python
matplotlib.pyplot.subplot(1,2,1) 
tmp_data[attri].value_counts(dropna = False).plot(kind='bar',figsize=(10,5))
matplotlib.pyplot.subplot(1,2,2) 
d[attri].value_counts(dropna = False).plot(kind='bar')
```

- 用最高频率值来填补缺失值

```python
mode = tmp_data[attri].mode()
#int(mode)
f = tmp_data[attri].fillna(int(mode))
print(f)
f.value_counts(dropna = False)
```

可视化对比

```python
matplotlib.pyplot.subplot(1,2,1) 
tmp_data[attri].value_counts(dropna = False).plot(kind='bar',figsize=(10,5))
matplotlib.pyplot.subplot(1,2,2) 
f.value_counts(dropna = False).plot(kind='bar',figsize=(10,5))
```



#### Beat属性缺失处理

可能是人为失误导致缺失值

- 将缺失部分剔除

```python
attri = "Beat"
d = tmp_data.dropna(subset=[attri])
d
```

可视化对比

```python
matplotlib.pyplot.subplot(2,1,1) 
tmp_data[attri].value_counts(dropna = False).plot(kind='bar',figsize=(15,8))
matplotlib.pyplot.subplot(2,1,2) 
d[attri].value_counts(dropna = False).plot(kind='bar',figsize=(15,8))
```
